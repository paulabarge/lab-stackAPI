#### Question 1: Find the questions and answers of last month.

```python
#To get the questions and answers of last month we have to convert todays date and today's last month date to a time stamp.
date_today = "07/07/2020 14:45:00"
date_amonth_ago = "07/06/2020 14:45:00"

#To change it, let's import time and date libraries
import time 
import datetime

#Let's now convert it to a timestamp using the time functions
ts_today = time.mktime(time.strptime(date_today, "%d/%m/%Y %H:%M:%S"))
ts_month_ago = time.mktime(time.strptime(date_amonth_ago, "%d/%m/%Y %H:%M:%S"))

print(ts_today)
print(ts_month_ago)
```

```python
#now that we have the info we need to ask the API for our questions and answers, let's obtain the info. 
SITE = StackAPI('stackoverflow') 
questions = SITE.fetch('questions', fromdate=1591533900, todate=1594125900)
questions
```

```python
#let's do the same for the answers 
SITE = StackAPI('stackoverflow') 
answers = SITE.fetch('answers', fromdate=1591533900, todate=1594125900)
answers
```

#### Question 2: Find the most voted question from last month till today, with at least a score of 5 and tagged with 'python'. 

```python
SITE = StackAPI('stackoverflow') 
mostvoted = SITE.fetch('questions', fromdate=1591533900, todate=1594125900, min=5, tag='python')
mostvoted
```
